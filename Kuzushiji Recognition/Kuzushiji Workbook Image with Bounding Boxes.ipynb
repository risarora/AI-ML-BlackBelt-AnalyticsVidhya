```python

def plotImage(img_name):
    image = plt.imread(path + img_name+'.jpg')
    fig = plt.figure(figsize=(15, 15))

    ax = fig.add_axes([0,0,1,1])
    plt.imshow(image)

    for _,row in data[data['FILE_NAME']==img_name].iterrows():
        file = '100241706_00006_1'
        charactersinImage = data[data['FILE_NAME']==file]['CHARACTER'].shape[0]


        xmin = row.X_min
        ymin = row.y_min
        width = row.length
        height = row.height
        edgecolor='r'
        rect = patches.Rectangle((xmin, ymin), width, height, edgecolor=edgecolor, facecolor='none')

        # add patch
        ax.add_patch(rect)

    print()
    plt.show()
```

```python
img_name = data['FILE_NAME'].unique()[0]
plotImage(img_name)
```
